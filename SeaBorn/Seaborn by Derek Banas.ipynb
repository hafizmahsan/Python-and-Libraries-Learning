```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
crash_df=sns.load_dataset('car_crashes')

crash_df
```

# Distribution Plots

## Distribution Plot

```python
sns.distplot(crash_df['not_distracted'])
```

```python
sns.distplot(crash_df['not_distracted'], kde=False)
```

```python
sns.distplot(crash_df['not_distracted'], kde=False, bins=25)
```

## Joint Plot

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

## Kde Plot (Kernal Density Estimation)

```python
sns.kdeplot(crash_df['alcohol'])
```

## Pair Plot

```python
sns.pairplot(crash_df)
```

```python
tips_df=sns.load_dataset('tips')
tips_df
```

```python
sns.pairplot(tips_df)
```

```python
sns.pairplot(tips_df, hue='sex')
```

```python
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

```python
sns.pairplot(tips_df, hue='sex', palette='bwr')
```

## Rug Plot

```python
sns.rugplot(tips_df['tip'])
```

## Styling

```python
sns.set_style('white')

plt.figure(figsize=(8, 4))

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('darkgrid')

plt.figure(figsize=(8, 4))

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('whitegrid')

plt.figure(figsize=(8, 4))

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('dark')

plt.figure(figsize=(8, 4))

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('ticks')

plt.figure(figsize=(8, 4))

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('ticks')

plt.figure(figsize=(8, 4))

sns.set_context('paper', font_scale=1.4)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('ticks')

plt.figure(figsize=(8, 4))

sns.set_context('talk', font_scale=1.4)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('ticks')

plt.figure(figsize=(8, 4))

sns.set_context('poster', font_scale=1.4)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('ticks')

plt.figure(figsize=(8, 4))

sns.set_context('paper', font_scale=1.5)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

sns.despine(left=True)
```

```python
sns.set_style('ticks')

plt.figure(figsize=(8, 4))

sns.set_context('paper', font_scale=1.5)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

sns.despine(left=True, bottom=True)
```
