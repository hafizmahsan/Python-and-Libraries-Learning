```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

plt.title('Our First Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})

#Adding xlabel and ylabel

plt.xlabel('X-Axis')

plt.ylabel('Y-Axis')

#xtickvalues and ytickvalues

plt.xticks([0, 1, 2, 3, 4, 5])

plt.yticks([0, 2, 4, 6, 8, 10])

plt.plot(x, y)

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

#Adding Label in Plot

plt.plot(x, y, label='2x', color='red')

plt.title('Our First Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')


plt.xticks([0, 1, 2, 3, 4, 5])

plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

#Adding Label in Plot

plt.plot(x, y, label='2x', color='red', linewidth=4)

plt.title('Our First Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')


plt.xticks([0, 1, 2, 3, 4, 5])

plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

#Adding Label in Plot

plt.plot(x, y, label='2x', color='red', linewidth=4, marker='.', markersize=10)

plt.title('Our First Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')


plt.xticks([0, 1, 2, 3, 4, 5])

plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

#Adding Label in Plot

plt.plot(x, y, label='2x', color='red', linewidth=4, marker='.', markersize=10, markeredgecolor='blue')

plt.title('Our First Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')


plt.xticks([0, 1, 2, 3, 4, 5])

plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

#Adding Label in Plot

plt.plot(x, y, label='2x', color='red', linewidth=4, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

plt.title('Our First Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')


plt.xticks([0, 1, 2, 3, 4, 5])

plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

#Adding Label in Plot

# plt.plot(x, y, label='2x', color='red', linewidth=4, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

#Short Hand Notation

#fmt = '[color][marker][line]'

plt.plot(x, y, 'r.--', label='2x')

plt.title('Our First Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')


plt.xticks([0, 1, 2, 3, 4, 5])

plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

#Adding Label in Plot

# plt.plot(x, y, label='2x', color='red', linewidth=4, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

#Short Hand Notation

#fmt = '[color][marker][line]'

plt.plot(x, y, 'r^--', label='2x')

plt.title('Our First Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')


plt.xticks([0, 1, 2, 3, 4, 5])

plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

#Adding Label in Plot

# plt.plot(x, y, label='2x', color='red', linewidth=4, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

#Short Hand Notation

#fmt = '[color][marker][line]'

plt.plot(x, y, 'b^--', label='2x')

#Line No. 2

x2 = np.arange(0, 4.5, 0.5)

plt.plot(x2, x2**2, 'r', label='x^2')

plt.title('Our First Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')


plt.xticks([0, 1, 2, 3, 4, 5])

plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

#Adding Label in Plot

# plt.plot(x, y, label='2x', color='red', linewidth=4, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

#Short Hand Notation

#fmt = '[color][marker][line]'

plt.plot(x, y, 'b^--', label='2x')

#Line No. 2

#Select Interval we want to plot points at

x2 = np.arange(0, 4.5, 0.5)

#Plot part of the Graph as Line
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')

#Plot remainder of the Graph as a Dot
plt.plot(x2[5:], x2[5:]**2, 'r--')


plt.title('Our First Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')


plt.xticks([0, 1, 2, 3, 4, 5])

plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

#Resizing your Graph

plt.figure(figsize=(5, 3), dpi=100)

#Adding Label in Plot

# plt.plot(x, y, label='2x', color='red', linewidth=4, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

#Short Hand Notation

#fmt = '[color][marker][line]'

plt.plot(x, y, 'b^--', label='2x')

#Line No. 2

x2 = np.arange(0, 4.5, 0.5)

plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')


plt.title('Our First Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')


plt.xticks([0, 1, 2, 3, 4, 5])

plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

#Resizing your Graph

plt.figure(figsize=(5, 3), dpi=100)

#Adding Label in Plot

# plt.plot(x, y, label='2x', color='red', linewidth=4, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

#Short Hand Notation

#fmt = '[color][marker][line]'

plt.plot(x, y, 'b^--', label='2x')

#Line No. 2

x2 = np.arange(0, 4.5, 0.5)

plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')


plt.title('Our First Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')


plt.xticks([0, 1, 2, 3, 4, 5])

plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

#Saving the Graph

plt.savefig('Data used in Learning Matplotlib Library\mygraph.png', dpi=300)

plt.show()
```

# Bar Chart

```python
labels = ['A', 'B', 'C', 'D']

values = [1, 2, 3, 4]

plt.bar(labels, values)

plt.show()
```

```python
labels = ['A', 'B', 'C']

values = [1, 4, 2]

plt.bar(labels, values)

plt.show()
```

```python
labels = ['A', 'B', 'C']

values = [1, 4, 2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')

plt.show()
```

```python
labels = ['A', 'B', 'C']

values = [1, 4, 2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')

plt.show()
```

```python
labels = ['A', 'B', 'C']

values = [1, 4, 2]

bars = plt.bar(labels, values)

#Another way of Doing Above Work

patterns = ['/', 'O', '*']

for bar in bars :
    bar.set_hatch(patterns.pop())

# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.show()
```

```python
labels = ['A', 'B', 'C']

values = [1, 4, 2]

bars = plt.bar(labels, values)

#Another way of Doing Above Work

patterns = ['/', 'O', '*']

for bar in bars :
    bar.set_hatch(patterns.pop(0))

# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.show()
```
